## AI Recommendation Systems

A recommender system is used by developers as a tool to suggest or recommend a particular item based on the item a user may have searched for from a large list of items. Recommendation engines make it quick and easy for a user to get recommended items because the items will be tailored to the users&#39; preferences due to the power of AI. The AI recommendation engine is a big tool that is used in the e-commerce sector. This is down to the fact that the recommender makes businesses more profitable, allows them to make more sales and revenues in general. There are two types of recommender systems, content based, and collaborative based.

**Content Based Recommender Systems –** The way that the content-based recommenders work is that the user provides the information for it to give the user the recommendation. Based on the users&#39; input that they provide, a profile for the user is generated. As the user makes more suggestions the user-based AI becomes more accurate. Information retrieval systems use the concepts of term frequency and inverse document frequency. Term frequency is the number of times a certain word is in a document and the inverse of document frequency based on the body of documents which is known as information document frequency. An example of IDF is if we search for &quot;the pink panther&quot; the recommender system knows that the word &quot;the&quot; will occur more frequently than &quot;pink&quot; or &quot;panther&quot; and so the AI knows that the important words in the search query are pink and panther.

**Collaborative Based Recommender Systems –** With collaborative recommender systems, it uses the similarities between the items and the user at the same time and that is how it is providing recommendations. This means that the recommender uses a filtering model that recommends an item to user B based on the similar interests of user A. This is known as serendipitous filtering. The system can automatically learn a user&#39;s interests without the need for engineering by hand for features.

```python
#Import Libraries


import numpy as np
import pandas as pd
import sklearn
```

```python


disney = pd.read_csv('disney_plus_titles.csv')
disney.columns = ['show_id', 'type', 'title', 'director', 'cast', 'country', 'date_added', 'release_year', 'rating', 'duration', 'listed_in', 'description']
disney.head()
```

```python
t = [15, 300, 160, 3.2]

x = disney.loc[:,('type', 'title', 'director')].values
x[0:5]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')
```

```python
disney['title'] = disney['title'].fillna('')
```

```python
tfidf_matrix = tfidf.fit_transform(disney['title'])
```

```python
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(disney.index, index=disney['title']).drop_duplicates()
```

```python
# Function that takes in movie and tv show title as input and outputs most similar movies and tv shows
def get_recommendations(name, cosine_sim=cosine_sim):
    # Get the index of the movies and tv shows that matches the title
    idx = indices[name]
    
    # Get the similar titles of all movies or tv shows with that movie or tv show
    sim = list(enumerate(cosine_sim[idx]))

    # Sort the movie or tv show based on the similarity title
    sim = sorted(sim, key=lambda x: x[1], reverse=True)

    #Get the titles of the 10 most similar movies or tv shows
    sim = sim[1:11]

    # Get the movie or tv show indices
    disney_indices = [i[0] for i in sim]

    # Return the top 10 most similar movies or tv shows
    return disney.iloc[disney_indices]
```

```python
get_recommendations('Star Wars: Visions')
```
